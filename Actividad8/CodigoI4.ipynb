```python
from numpy import *
import pylab as p
import numpy as np
# Definition of parameters
a=1.2
w=(2.0*np.pi)/10.0
mu= 8.53
def dX_dt(X, t=0):
    xp=X[1]
    yp=mu*(1-X[0]**2)*X[1]-X[0]+a*np.sin(w*t)
    return array([xp, yp])
```

```python
from scipy import integrate
t  = linspace(300, 600, 1200)              # time
X0 = array([-1.0, 0.0])                     # initials conditions: 10 rabbits and 5 foxes
X, infodict = integrate.odeint(dX_dt, X0, t, full_output=True)
infodict['message']  
```

```python
from numpy import loadtxt
from pylab import figure, plot, xlabel, grid, hold, legend, title, savefig, xlim, ylabel, ylim
from matplotlib.font_manager import FontProperties
%matplotlib inline
x, y = X.T

f1=p.figure(figsize=(15,3))
plot(t, x, 'hotpink')
grid()
xlim(300,600)
ylim(-3,3)
xlabel('tiempo')
ylabel('x')
title('Posici√≥n contra tiempo')
savefig('Posicion_contra_tiempo2.png')
```
