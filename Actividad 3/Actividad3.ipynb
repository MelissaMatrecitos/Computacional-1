```python
# Cargar a la memoria de trabajo las bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Leer el arcivo Diciembre
dfDic = pd.read_csv('LindenbergDic.txt', skiprows=2, sep='\s+')
```

```python
dfDic = dfDic.drop(dfDic.index[[0,1]])
```

```python
dfDic
```

```python
dfDic=dfDic[:-30]
```

```python
dfDic.tail()
```

```python
#Leer el arcivo Junio
dfJun = pd.read_csv('LindenbergJun.txt', skiprows=2, sep='\s+')
```

```python
dfJun.head()
```

```python
dfJun = dfJun.drop(dfJun.index[[0,1]])
```

```python
dfJun
```

```python
dfJun=dfJun[:-27]
```

```python
dfJun.tail()
```

```python
cols = ['PRES', 'HGHT', 'TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
dfJun[cols] = dfJun[cols].apply(pd.to_numeric, errors='coerce', axis=1)
dfDic[cols] = dfDic[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
dfJun.dtypes
```

```python
dfDic.dtypes
```

```python
#Gráfica de la variación de la presión con respecto a la altura (Junio)
JPRES=dfJun['PRES']
JHGHT=dfJun['HGHT']
plt.plot(JPRES, JHGHT, "hotpink", label='Presión'); plt.legend(loc='best')
plt.title('Presión con Respecto a la Altura (Junio)')
plt.ylabel('Altura (m)')
plt.xlabel('Presión  (hPa)')
plt.grid(True)
plt.show()

#Gráfica de la variación de la presión con respecto a la altura (Diciembre)
DPRES=dfDic['PRES']
DHGHT=dfDic['HGHT']
plt.plot(DPRES, DHGHT, "darkturquoise", label='Presión'); plt.legend(loc='best')
plt.title('Presión con Respecto a la Altura (Diciembre)')
plt.ylabel('Altura (m)')
plt.xlabel('Presión  (hPa)')
plt.grid(True)
plt.show()
```

```python
#Gráfica de la variación de la temperatura con respecto a la altura (Junio)
JTEMP=dfJun['TEMP']
JHGHT=dfJun['HGHT']
plt.plot(JTEMP, JHGHT, "hotpink", label='Temperatura'); plt.legend(loc='best')
plt.title('Temperatura con Respecto a la Altura (Junio)')
plt.ylabel('Altura (m)')
plt.xlabel('Temperatura (C)')
plt.grid(True)
plt.show()

#Gráfica de la variación de la temperatura con respecto a la altura (Diciembre)
DTEMP=dfDic['TEMP']
DHGHT=dfDic['HGHT']
plt.plot(DTEMP, DHGHT, "darkturquoise", label='Temperatura'); plt.legend(loc='best')
plt.title('Temperatura con Respecto a la Altura (Diciembre)')
plt.ylabel('Altura (m)')
plt.xlabel('Temperatura (C)')
plt.grid(True)
plt.show()
```

```python
#Gráfica de la variación de la temperatura y temperatura de rocío con respecto a la altura (Junio)
JTEMP=dfJun['TEMP']
JTEMPR=dfJun['DWPT']
JHGHT=dfJun['HGHT']
plt.plot(JTEMP, JHGHT, "hotpink", label='Temperatura')
plt.plot(JTEMPR, JHGHT, "purple", label='Temperatura de rocío')
plt.legend(loc='best')
plt.title('Temperatura y Temperatura de rocío con Respecto a la Altura (Junio)')
plt.ylabel('Altura (m)')
plt.xlabel('Temperatura (C)')
plt.grid(True)
plt.show()

#Gráfica de la variación de la temperatura y temperatura de rocío con respecto a la altura (Diciembre)
DTEMP=dfDic['TEMP']
DTEMPR=dfDic['DWPT']
DHGHT=dfDic['HGHT']
plt.plot(DTEMP, DHGHT, "darkturquoise", label='Temperatura')
plt.plot(DTEMPR, DHGHT, "navy", label='Temperatura de Rocío')
plt.legend(loc='best')
plt.title('Temperatura y Temperatura de rocío con Respecto a la Altura (Diciembre)')
plt.ylabel('Altura (m)')
plt.xlabel('Temperatura (C)')
plt.grid(True)
plt.show()
```

```python
#Gráfica de la rapidez de los vientos en nudos contra altura (Junio)
JRAP=dfJun['SKNT']
JHGHT=dfJun['HGHT']
plt.plot(JRAP, JHGHT, "hotpink", label='Rapidez'); plt.legend(loc='best')
plt.title('Rapidez de los vientos en nudos (Junio)')
plt.xlabel('Rapidez')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()

#Gráfica de la rapidez de los vientos en nudos contra la altura (Diciembre)
DRAP=dfDic['SKNT']
DHGHT=dfDic['HGHT']
plt.plot(DRAP,DHGHT, "darkturquoise", label='Rapidez'); plt.legend(loc='best')
plt.title('Rapidez de los vientos en nudos (Diciembre)')
plt.ylim(0,27000)
plt.xlim(0,100)
plt.xlabel('Rapidez')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gráfica de la humedad relativa con respecto a la altura (Junio)
JHUMR=dfJun['RELH']
JHGHT=dfJun['HGHT']
plt.plot(JHUMR, JHGHT, "hotpink", label='Humedad Relativa'); plt.legend(loc='best')
plt.title('Humedad Relativa con Respecto a la Altura (Junio)')
plt.ylabel('Altura (m)')
plt.xlabel('Humedad Relativa (%)')
plt.grid(True)
plt.show()

#Gráfica de la humedad relativa con respecto a la altura (Diciembre)
DHUMR=dfDic['RELH']
DHGHT=dfDic['HGHT']
plt.plot(DHUMR, DHGHT, "darkturquoise", label='Humedad Relativa'); plt.legend(loc='best')
plt.title('Humedad Relativa con Respecto a la Altura (Diciembre)')
plt.ylim(0,27138)
plt.ylabel('Altura (m)')
plt.xlabel('Humedad Relativa (%)')
plt.grid(True)
plt.show()
```
